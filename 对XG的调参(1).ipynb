```python
def get_random_data(n):
        mt_x_1 = np.random.rand(n,5)
        mt_x_b1 = np.random.binomial(1, 0.4,size = n).reshape(-1,1)
        mt_x_b2 = np.random.binomial(1, 0.8,size = n).reshape(-1,1)
        mt_x = np.concatenate((mt_x_1, mt_x_b1,mt_x_b2),axis = 1)
        f1= np.zeros(n).reshape(n,1)
        f2= np.zeros(n).reshape(n,1)
        f3= np.zeros(n).reshape(n,1)
        f4= np.zeros(n).reshape(n,1)
        f5= np.zeros(n).reshape(n,1)
        f6= np.zeros(n).reshape(n,1)
        f7= np.zeros(n).reshape(n,1)
        for i in range(n):
            f1[i]=6*mt_x[i,0]
            f2[i]=-3*math.cos((6*mt_x[i,1]-5)*math.pi/3)
            f3[i]=((2-mt_x[i,2])**2)/5
            f4[i]=0
            f5[i]=0
            f6[i]=5*mt_x[i,5]
            f7[i]=-7*mt_x[i,6]   
        mt_y = np.zeros(n).reshape(n,1)
        mt_y = f1+f2+f3+f4+f5+f6+f7+np.random.normal(0, 3, n ).reshape(n,1)
        mt_data = np.concatenate((mt_x, mt_y),axis=1)
        mt_data = pd.DataFrame(mt_data)
        return mt_data
# random_100 = get_random_data(100)
# random_800 = get_random_data(800)
# random_2000 = get_random_data(2000)
```

```python
def get_random_more(n):
    mt_x_1 = np.random.rand(n,5)
    mt_x_integar = np.random.choice([18,20,22,23,24,26], size = n,p=[0.1,0.2,0.3,0.1,0.1,0.2]).reshape(-1,1)
    mt_x_b1 = np.random.binomial(1, 0.4,size = n).reshape(-1,1)
    mt_x_tri1 = np.random.binomial(1, 0.5,size = n).reshape(-1,1)
    mt_x_tri2 = np.random.binomial(1, 0.8,size = n).reshape(-1,1)
    mt_x = np.concatenate((mt_x_1, mt_x_integar,mt_x_b1,mt_x_tri1,mt_x_tri2),axis = 1)
    mt_y = np.zeros(n).reshape(n,1)
    f1= np.zeros(n).reshape(n,1)
    f2= np.zeros(n).reshape(n,1)
    f3= np.zeros(n).reshape(n,1)
    f4= np.zeros(n).reshape(n,1)
    f5= np.zeros(n).reshape(n,1)
    f6= np.zeros(n).reshape(n,1)
    f7= np.zeros(n).reshape(n,1)
    f8= np.zeros(n).reshape(n,1)
    f9= np.zeros(n).reshape(n,1)
    for i in range(n):
        f1[i]=6*mt_x[i,0]
        f2[i]=-3*math.cos((6*mt_x[i,1]-5)*math.pi/3)
        f3[i]=((2-mt_x[i,2])**2)/5
        f4[i]=0
        f5[i]=0
        f6[i]=0.5*mt_x[i,5]
        f7[i]=-7*mt_x[i,6] 
        f8[i]=5*mt_x[i,7]
        f9[i]=-4*mt_x[i,8]   
    mt_y = np.zeros(n).reshape(n,1)
    mt_y = f1+f2+f3+f4+f5+f6+f7+f8+f9+np.random.normal(0, 1, n ).reshape(n,1)
    mt_data = np.concatenate((mt_x, mt_y),axis=1)
    mt_data = pd.DataFrame(mt_data)
    for i in range(len(mt_data)):
        if mt_data.iloc[i,7] + mt_data.iloc[i,8] > 1:
            mt_data.iloc[i,8] = 0
    return mt_data
# random_more_100 = get_random_more(100)
# random_more_800 = get_random_more(800)
# random_more_2000 = get_random_more(2000)
```
